## 2.Pra que serve cada um deles?

###### n_estimators:
O numero de arvores na floresta

###### criterion: {“gini”, “entropy”, “log_loss”}
A função para medir a qualidade de uma divisão. Os critérios suportados são “gini” para a impureza de Gini e “log_loss” e “entropy” para o Ganho de informação da Shannon, veja Formulação matemática. Nota: Este parâmetro é específico para árvore.

###### max_depth:
Profundidade máxima da arvore

###### min_samples_split:
O número mínimo de amostras necessárias para dividir um nó interno:

Se int, considere min_samples_split como o número mínimo.

Se float, então min_samples_split é uma fração e são os mínimos número de amostras para cada divisão.ceil(min_samples_split * n_samples)

###### min_samples_leaf:
O número mínimo de amostras necessárias para estar em um nó de folha. Um ponto de divisão a qualquer profundidade só será considerado se sair em pelo menos min_samples_leaf amostras de treinamento em cada uma das esquerdas e galhos certos. Isso pode ter o efeito de suavizar o modelo, especialmente em regressão.

###### max_features{“sqrt”, “log2”, None}:
O número de recursos a serem considerados ao procurar a melhor divisão:

###### bootstrap:
Se as amostras de bootstrap são usadas na construção de árvores. Se Falso, o todo o conjunto de dados é usado para construir cada árvore.

###### oob_score:
Se usar amostras fora do saco para estimar a pontuação de generalização. Por padrão, accuracy_score é usado

###### random_state:
Controla a aleatoriedade do bootstrapping das amostras utilizadas ao construir árvores 

###### ccp_alpha:
Parâmetro de complexidade usado para Poda de Complexidade de Custo Mínimo. O subárvore com a maior complexidade de custo que é menor do que ccp_alpha será escolhido. Por padrão, nenhuma poda é realizadaárvore. Se Nenhum, os nós serão expandidos até todas as folhas são puras ou até que todas as folhas contenham menos de min_samples_split amostras.
