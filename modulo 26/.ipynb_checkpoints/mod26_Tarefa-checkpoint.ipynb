# Tarefa - Álgebra Linear

A célula abaixo gera dados de notas de duas provas, p1 e p2, conforme em aula, mas com mais alunos.

## 1. Reduzindo correlações

a) Já vimos que expressar os dados através da média e diferença é uma transformação linear. Obtenha a média e a diferença através de uma multiplicação matricial, conforme visto em aula.

b) Calcule a correlação entre os dados da **p1** e **p2**

c) Calcule a correlação entre os dados transformados (média e diferença)

d) Comente as correlações observadas.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
```

###### a) Já vimos que expressar os dados através da média e diferença é uma transformação linear. Obtenha a média e a diferença através de uma multiplicação matricial, conforme visto em aula.

```python
m = np.array([[.5,.5],   
              [-1, 1]])  

df['média'], df['diferença'] = np.matmul(m, df[['p1', 'p2']].to_numpy().transpose())

(df.style 
   .format(precision=2,  
           decimal=','))
```

```python

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

# Gráfico dos valores de p1 na horizontal X p2 na vertical
scatter1 = ax[0].scatter(x=df.p1, y=df.p2)
ax[0].set_xlabel('p1')  
ax[0].set_ylabel('p2')  

# Gráfico da Média em relação a Diferença de p1 X p2
scatter2 = ax[1].scatter(x=df['média'], y=df['diferença'])
ax[1].axhline(y=0, linewidth=0.5, linestyle='--')  
ax[1].set_xlabel('Média') 
ax[1].set_ylabel('Diferença') 

plt.show()
```

###### b) Calcule a correlação entre os dados da p1 e p2

```python
print("Correlação entre p1 e p2:", df.p1.corr(df.p2))
```

###### c) Calcule a correlação entre os dados transformados (média e diferença)

```python
print("Correlação entre a média e a diferença:", df['média'].corr(df['diferença']))
```

###### d) Comente as correlações observadas.

<strong>Correlação entre os dados das variáveis p1 e p2:</strong>

- A correlação de 0,89 demonstra uma forte correlação positiva entre as amostras das colunas p1 e p2. Isso significa que essas variáveis têm uma relação positiva, indicando uma forte associação e semelhança entre os dois conjuntos de dados.

<strong>Correlação entre os dados transformados (média e diferença):</strong>

- O valor resultante de -0,13 indica uma correlação negativa relativamente baixa entre as variáveis transformadas. Isso significa que há uma relação inversa, ou seja, à medida que uma variável aumenta, a outra tende a diminuir, e vice-versa. No entanto, o valor próximo de zero (-0,13) indica uma correlação inversa fraca. Essa relação inversa pode não ser tão significativa ou previsível.


## 2. Soma e diferença

Realize agora uma transformação semelhante, mas que ao invés da média, dê a soma da **p1** e **p2**. Construa o gráfico e conclua que diferenças visuais você observa.

```python
df['soma'] = df.p1 + df.p2 

(df.style 
   .format(precision=2, 
           decimal=','))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# Gráfico dos valores de p1 na horizontal X p2 na vertical
scatter1 = ax[0].scatter(x=df.p1, y=df.p2, c=df['média'])
ax[0].set_xlabel('p1')  
ax[0].set_ylabel('p2')  

# Barra de cores para representar a média
cbar = fig.colorbar(scatter1, ax=ax[0])
cbar.set_label('Média')  

# Gráfico da Soma em relação a Diferença de p1 X p2
scatter2 = ax[1].scatter(x=df['soma'], y=df['diferença'], c=df['média'])
ax[1].axhline(y=0, linewidth=0.5, linestyle='--')  
ax[1].set_xlabel('Soma')  
ax[1].set_ylabel('Diferença')  

plt.show()
```

<blockquote>O primeiro gráfico mostra uma distribuição proporcional entre os valores de 0 a 10 tanto na variável p1 quanto na variável p2, indicando uma distribuição equilibrada e sem concentração em uma faixa específica. A média resultante reflete essa distribuição.

Após a transformação dos dados para gerar a soma e a diferença entre p1 e p2, observa-se que a variação é proporcional em todos os dados, sugerindo que as operações de soma e diferença preservam a proporção entre os valores das variáveis.

Essas observações destacam a característica de proporcionalidade entre p1 e p2, tanto na distribuição original quanto nas operações de soma e diferença, o que é importante para análise e interpretação dos dados. Além disso, a relação adicional entre a soma, a média e as variáveis p1 e p2 reforça a importância da proporção na análise dos dados.</blockquote>
