# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc_raw.groupby('munResNome')['IDADEMAE', 'IDADEPAI'].mean()
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
filtrado = sinasc_raw[sinasc_raw['DTNASC']== '2019-08-05']

filtrado.groupby(['ESCMAE', 'SEXO'])['PESO'].mean()
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc = sinasc_raw[sinasc_raw['munResNome'] != 'Município ignorado - RO']
```

```python
sinasc.groupby(['munResNome'])['IDADEMAE'].count().idxmin()
```

```python
municipio = sinasc[sinasc['munResNome']== 'Castanheiras']
```

```python
municipio.groupby('munResNome')['IDADEMAE', 'IDADEPAI'].agg(['min','mean','max'])
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
sinasc['DTNASC'] = pd.to_datetime(sinasc['DTNASC'], format='%d-%m-%Y')
meses = sinasc[sinasc['DTNASC'].dt.month == 3]
meses.groupby('munResNome')['IDADEMAE'].count().idxmax()
```

```python
municipio_pv = sinasc[sinasc['munResNome']== 'Porto Velho']
```

```python
municipio_pv.groupby('munResNome')['QTDFILVIVO'].agg(['min', 'mean', 'max'])
```

```python
municipio_pv.groupby('munResNome')['IDADEPAI'].agg(['min', 'mean', 'max'])
```
