````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv('previsao_renda.csv')
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
df2 = df.select_dtypes(include=('object'))
````

````python
fig, axes = plt.subplots(1, 2, figsize=(12,4))
sns.countplot(x='data_ref', hue='sexo', data=df2, ax=axes[0])
sns.countplot(x='data_ref', hue='tipo_renda', data=df2, ax=axes[1])
for ax in axes:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
````

````python
fig, axes = plt.subplots(1, 2, figsize=(12,4))
sns.countplot(x='data_ref', hue='estado_civil', data=df2, ax=axes[0])
sns.countplot(x='data_ref', hue='tipo_residencia', data=df2, ax=axes[1])
for ax in axes:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
````

### Não houve mudanças significativas nas variaveis qualitativas ao longo do tempo

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
fig, axes = plt.subplots(1, 2, figsize=(12,4))
graf = df2.groupby(['sexo', 'data_ref']).size().unstack().mean()
graf2 = df2.groupby(['tipo_renda', 'data_ref']).size().unstack().mean()

axes[0].bar(graf.index, graf.values)
axes[0].set_xlabel('sexo')
axes[1].bar(graf2.index, graf2.values)
axes[1].set_xlabel('tipo_renda')

for ax in axes:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
````

````python
fig, axes = plt.subplots(1, 2, figsize=(12,4))
graf = df2.groupby(['estado_civil', 'data_ref']).size().unstack().mean()
graf2 = df2.groupby(['tipo_residencia', 'data_ref']).size().unstack().mean()

axes[0].bar(graf.index, graf.values)
axes[0].set_xlabel('Estado civil')
axes[1].bar(graf2.index, graf2.values)
axes[1].set_xlabel('Tipo residencia')

for ax in axes:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
````
