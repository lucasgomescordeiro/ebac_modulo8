````python
import pandas as pd
import seaborn as sns
````

````python
df = pd.read_csv('previsao_renda.csv')
````

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
df2 = df[['posse_de_veiculo', 'posse_de_imovel', 'renda']]
````

````python
sns.clustermap(df2.corr(), figsize=[5, 5]);
````

````python
sns.pointplot(x="posse_de_imovel", y="renda", dodge=True, color='orange', errorbar=('ci', 95), data=df2);
````

````python
sns.pointplot(x="posse_de_veiculo", y="renda", dodge=True, color='blue', errorbar=('ci', 95), data=df2);
````

Posse de veiculo parece dicriminar melhor a renda, com um intervalo de confiança menor indica uma estimativa mais precisa.
Quem possui veiculo tem uma renda maior
